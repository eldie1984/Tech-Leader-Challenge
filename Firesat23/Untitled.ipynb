```python
def row_to_dict(row:list,header:list)->dict:
    a_dict={}
    for i in range(len(row)):
        a_dict[header[i]]=row[i]
    a_dict['continent']=get_continent(a_dict['x'], a_dict['y'])
    return(a_dict)
```
